

## ðŸŸ¢ **Beginner: The Basics of Lists**

1. **What is a list?**

   * A list is an ordered, mutable collection of items.
   * Example: `my_list = [1, 2, 3, 4]`

2. **Creating a list**

   * Empty list: `[]` or `list()`
   * With values: `[1, 2, 'a', True]`

3. **Accessing elements**

   * By index: `my_list[0]` â†’ `1`
   * Negative index: `my_list[-1]` â†’ last element

4. **List length**

   * `len(my_list)`

5. **Slicing**

   * `my_list[1:3]` â†’ returns elements at index 1 and 2
   * `my_list[:2]`, `my_list[::2]`

6. **Modifying values**

   * `my_list[0] = 100`

7. **Adding elements**

   * `append(x)` â†’ add at end
   * `insert(i, x)` â†’ insert at position `i`
   * `extend([x, y])` â†’ add multiple items

8. **Removing elements**

   * `remove(x)` â†’ removes first occurrence of `x`
   * `pop()` â†’ removes and returns last item
   * `pop(i)` â†’ removes item at index `i`
   * `clear()` â†’ empties the list

## ðŸŸ¡ **Intermediate: More Useful List Operations**

9. **Looping through lists**

   * `for item in my_list:`
   * `for i in range(len(my_list)):` (for index-based access)

10. **Checking membership**

    * `x in my_list`, `x not in my_list`

11. **List comprehension**

    * `[x**2 for x in range(5)]` â†’ `[0, 1, 4, 9, 16]`

12. **Nested lists**

    * `matrix = [[1,2], [3,4]]`
    * Access: `matrix[1][0]` â†’ `3`

13. **Sorting**

    * `sort()` (in-place)
    * `sorted(my_list)` (returns new sorted list)
    * `my_list.sort(reverse=True)` â†’ descending order

14. **Copying lists**

    * `copy = my_list[:]`
    * `copy = my_list.copy()`

15. **List methods**

    * `index(x)` â†’ returns index of x
    * `count(x)` â†’ counts occurrences of x
    * `reverse()` â†’ reverses list in-place

## ðŸ”µ **Advanced: Pro-Level List Handling**

16. **Unpacking**

    * `a, b, c = [1, 2, 3]`
    * `a, *rest = [1, 2, 3, 4]`

17. **Enumerate**

    * `for index, value in enumerate(my_list):`

18. **Zip**

    * Combine multiple lists: `list(zip(list1, list2))`

19. **List flattening**

    * `flat = [item for sublist in nested for item in sublist]`

20. **Using `map()` and `filter()`**

    * `map`: `list(map(str.upper, ['a', 'b']))`
    * `filter`: `list(filter(lambda x: x > 0, [-1, 2, 3]))`

21. **List vs Tuple**

    * Lists are mutable, tuples are immutable (`[]` vs `()`)

22. **Memory efficiency & performance**

    * Use list comprehension over loops for speed
    * Avoid growing lists in loops â€” use `extend()` or comprehensions
