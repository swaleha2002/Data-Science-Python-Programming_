

## ðŸŸ¢ **Beginner: Understanding Sets**

1. **What is a set?**

   * A **set** is an **unordered**, **mutable** collection of **unique** elements.
   * Example: `my_set = {1, 2, 3}`

2. **Creating sets**

   * With `{}`: `s = {1, 2, 3}`
   * From list/tuple: `s = set([1, 2, 3])`
   * Empty set: `set()` (not `{}`, which creates a dictionary)

3. **Properties of sets**

   * No duplicates: `set([1, 2, 2, 3]) â†’ {1, 2, 3}`
   * Elements must be **hashable** (immutable types only)

4. **Accessing elements**

   * You **cannot access by index** (sets are unordered)
   * Use loops:

     ```python
     for item in my_set:
         print(item)
     ```

5. **Check membership**

   * `'a' in my_set` â†’ `True` or `False`

---

## ðŸŸ¡ **Intermediate: Modifying Sets**

6. **Adding elements**

   * `add(x)` â†’ adds `x` if not present

7. **Removing elements**

   * `remove(x)` â†’ removes `x`; raises `KeyError` if not found
   * `discard(x)` â†’ removes `x` if present; does nothing if not
   * `pop()` â†’ removes and returns a random element
   * `clear()` â†’ empties the set

8. **Set length**

   * `len(my_set)`

9. **Copying sets**

   * `s2 = s1.copy()`

10. **Converting between types**

* Set â†’ List: `list(my_set)`
* List â†’ Set: `set(my_list)` (removes duplicates)

---

## ðŸ”µ **Advanced: Set Operations (Math-like)**

11. **Union**

* `a | b` or `a.union(b)`
* All unique elements from both sets

12. **Intersection**

* `a & b` or `a.intersection(b)`
* Elements common to both sets

13. **Difference**

* `a - b` or `a.difference(b)`
* Elements in `a` but not in `b`

14. **Symmetric Difference**

* `a ^ b` or `a.symmetric_difference(b)`
* Elements in either set, but **not both**

15. **Update operations**

* `a.update(b)` â†’ modifies `a` with `a | b`
* `a.intersection_update(b)` â†’ modifies `a` with `a & b`

16. **Set comparisons**

* `a == b` â†’ True if sets are equal
* `a.issubset(b)` or `a <= b`
* `a.issuperset(b)` or `a >= b`
* `a.isdisjoint(b)` â†’ True if no common elements

## ðŸŸ£ **Specialized Use & Tips**

17. **Use sets to remove duplicates**

```python
nums = [1, 2, 2, 3]
unique = list(set(nums))  # [1, 2, 3]
```

18. **Faster membership checks**

* `x in set` is **O(1)** time on average (much faster than lists)

19. **Frozen sets (immutable sets)**

* `fs = frozenset([1, 2, 3])`
* Can't add/remove elements
* Can be used as keys in dictionaries or elements in other sets

20. **Set comprehension**

* `{x for x in range(10) if x % 2 == 0}`
